matplotlib:

Useful for making Plots



```python
#importing matplotlib libraray
import matplotlib.pyplot as plt
```

```python
#importing numpy to get for our plots
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the data

```python
#sine wave
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
#adding the title x-axis and y-axis lable
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel('Sine values')
plt.title("Sine Wave")
plt.show()
```

```python
#parabola
y=np.linspace(-10,10,20)
x=4*y*y
plt.plot(x,y)

plt.show()
```

```python
plt.plot(x,y,"r+")
plt.show()
```

```python
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

Bar Plot

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=["English","French","Spanish","Latin","German"]
people=[100,50,150,40,70]
ax.bar(languages,people)
plt.xlabel("Languages")
plt.ylabel("Number of people")
plt.show()
```

Pie chart

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
languages=["English","French","Spanish","Latin","German"]
people=[100,50,150,40,70]
ax.pie(people,labels=languages,autopct="%1.1f%%")
plt.show()
```

Scatter Plot

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color="g")
ax.scatter(x,z,color="r")
plt.show()
```

3rd scatter plot

```python
fig3=plt.figure()
ax=plt.axes(projection="3d")
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap="Blues")
plt.show()
```
